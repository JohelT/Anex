```python
import os
import io
import pandas as pd
import numpy as np
import spacy as sp
from spacy import displacy
from spacy.tokens import SpanGroup
```

```python
#Exctracting the student responses
fp = os.path.join('data', 'stevens_laptop.csv')
laptop_df = pd.read_csv(fp)
laptop_df
nlp = sp.load("en_core_web_sm")
```

```python
anddf = laptop_df[laptop_df.Explanation.str.contains('and.+and.+and.+and').fillna(False)]
minlen = anddf.Explanation.str.len().min()
minlen
```

```python
keep_responses = anddf[anddf.Explanation.str.len() <= 70]
keep_responses
```

```python
response_text_1 = keep_responses.iloc[0].Explanation
doc = nlp(response_text_1)
displacy.serve(doc, style="dep")
```

```python
response_text_2 = keep_responses.iloc[1].Explanation
doc1 = nlp(response_text_2)
displacy.serve(doc1, style="dep")
```
